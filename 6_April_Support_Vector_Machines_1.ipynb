##Q1. What is the mathematical formula for a linear SVM?

##Ans:---

###A linear Support Vector Machine (SVM) is a type of binary classifier that finds the hyperplane that best separates the data points of different classes. The mathematical formula for a linear SVM can be expressed as follows:

``` 
Given a set of training data {(x1, y1), (x2, y2), ..., (xn, yn)}, 
where xi represents the feature vector for the ith training example and yi represents its corresponding class label (yi = +1 or -1), 
we seek to find the hyperplane that maximizes the margin between the two classes.

The hyperplane is represented by the equation:

w^T x + b = 0

where w is the weight vector and b is the bias term.

The margin is given by:

2/||w||

where ||w|| is the Euclidean norm of the weight vector w.

The optimization problem for finding the optimal hyperplane can be formulated as:

minimize: 1/2 ||w||^2

subject to: yi(w^T xi + b) >= 1, for i = 1,2,...,n
```
###This is a constrained optimization problem, where the objective is to minimize the norm of the weight vector, subject to the constraint that all data points are correctly classified with a margin of at least 1. This is equivalent to maximizing the margin between the two classes.

###The solution to this optimization problem can be obtained using various numerical methods, such as gradient descent or quadratic programming. Once the optimal values of w and b are obtained, the hyperplane can be used to classify new data points based on their feature vectors.

##Q2. What is the objective function of a linear SVM?

##Ans:--

###The objective function of a linear Support Vector Machine (SVM) is to find the hyperplane that maximizes the margin between the two classes while correctly classifying all training examples. This objective function can be expressed as a constrained optimization problem.

###The optimization problem for a linear SVM is to minimize 1/2 ||w||^2 subject to the constraint that yi(w^T xi + b) >= 1 for all i, where xi is the ith training example, yi is its corresponding class label (+1 or -1), w is the weight vector, and b is the bias term.

###The objective function consists of two parts. The first part, 1/2 ||w||^2, is the regularization term that penalizes the norm of the weight vector to prevent overfitting. The second part is the hinge loss function, which is used to ensure that all training examples are correctly classified with a margin of at least 1. The hinge loss function is given by max(0, 1 - yi(w^T xi + b)).

###The goal of the optimization problem is to find the values of w and b that minimize the objective function while satisfying the constraint. The solution to this optimization problem corresponds to the hyperplane that maximizes the margin between the two classes.

##Q3. What is the kernel trick in SVM?

##Ans:--

###The kernel trick is a technique used in Support Vector Machines (SVMs) that allows for nonlinear classification by mapping the input data into a higher-dimensional space.

###The basic idea of the kernel trick is to replace the dot product in the feature space with a non-linear kernel function that computes the inner product of the data points in a transformed space. This enables the SVM to find a nonlinear decision boundary that separates the data points of different classes in the higher-dimensional space.

###Instead of explicitly mapping the data into the higher-dimensional space, which can be computationally expensive or even impossible for very high-dimensional spaces, the kernel function computes the inner product of the data points in the transformed space without actually transforming them. This is done by defining a kernel function that corresponds to a specific transformation of the input data into the higher-dimensional space.

###Some commonly used kernel functions include the linear kernel, polynomial kernel, Gaussian radial basis function (RBF) kernel, and sigmoid kernel. The choice of kernel function depends on the characteristics of the data and the problem at hand.

###The kernel trick makes it possible to use SVMs for nonlinear classification problems, where linear separation is not possible in the input space. By using a kernel function, the SVM can implicitly learn a nonlinear decision boundary in the higher-dimensional space without actually computing the mapping. This can lead to improved performance and better generalization on complex datasets.

##Q4. What is the role of support vectors in SVM Explain with example

##Ans:---

###In Support Vector Machines (SVMs), support vectors are the data points that lie closest to the decision boundary or hyperplane. The role of support vectors is crucial in SVMs as they determine the position and orientation of the decision boundary and are used to define the margin of the classifier.

###In an SVM, the objective is to find the hyperplane that maximizes the margin between the two classes while correctly classifying all training examples. The margin is the distance between the decision boundary and the closest data points of each class. The support vectors are the data points that lie on the margin, or those that would change the position of the decision boundary if removed from the dataset.

###Consider a binary classification problem with two classes, labeled +1 and -1, and a two-dimensional feature space. ###In this example, the hyperplane is a straight line that separates the positive and negative classes, with the equation y = x.

###The margin of the classifier is defined by the closest data points to the hyperplane. In this case, the closest data points are (1, 1) and (3, 3), which belong to the positive and negative classes, respectively.

###These two data points are the support vectors, which define the margin of the classifier. The distance between the hyperplane and the support vectors is the margin.

##Q5. Illustrate with examples and graphs of Hyperplane, Marginal plane, Soft margin and Hard margin in SVM?

###Support Vector Machine (SVM) is a popular machine learning algorithm used for classification and regression analysis. SVM tries to find the hyperplane that best separates the data points into different classes. Here are some examples and graphs that illustrate the different concepts of SVM:

* Hyperplane: A hyperplane is a decision boundary that separates the data points of one class from another. In two dimensions, a hyperplane is simply a line, while in higher dimensions, it is a plane. The equation for a hyperplane in two dimensions is given by:

```
y = wx + b

where 
w is the weight vector, 
x is the input vector, 
b is the bias, 
and y is the output.

```
###Marginal plane: A marginal plane is the hyperplane that is closest to the data points of both classes. It is also known as the maximum margin hyperplane. The distance between the marginal plane and the closest data point of either class is known as the margin. SVM tries to maximize this margin to find the best hyperplane.

##Q6. SVM Implementation through Iris dataset.
```
~ Load the iris dataset from the scikit-learn library and split it into a training set and a testing setl
~ Train a linear SVM classifier on the training set and predict the labels for the testing setl
~ Compute the accuracy of the model on the testing setl
~ Plot the decision boundaries of the trained model using two of the featuresl
~ Try different values of the regularisation parameter C and see how it affects the performance of
the model.
```

````python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import numpy as np

# Load the iris dataset
iris = load_iris()
X = iris.data[:, :2]  # we only take the first two features
y = iris.target

# Split the data into a training set and a testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

# Train a linear SVM classifier on the training set
clf = LinearSVC(random_state=0, C=1)
clf.fit(X_train, y_train)

# Predict the labels for the testing set
y_pred = clf.predict(X_test)

# Compute the accuracy of the model on the testing set
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Plot the decision boundaries of the trained model using two of the features
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02), np.arange(y_min, y_max, 0.02))
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.xticks(())
plt.yticks(())
plt.title('Linear SVM Decision Boundaries')

# Try different values of the regularization parameter C and see how it affects the performance of the model
for C in [0.1, 1, 10]:
    clf = LinearSVC(random_state=0, C=C)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print("C =", C, "Accuracy:", accuracy)
````

###The code first loads the iris dataset and splits it into a training set and a testing set. It then trains a linear SVM classifier on the training set and predicts the labels for the testing set. The accuracy of the model on the testing set is computed and printed.

###The code then plots the decision boundaries of the trained model using two of the features (sepal length and sepal width).

###Finally, the code tries different values of the regularization parameter C and computes the accuracy of the model on the testing set for each value of C. The results show that changing the value of C does not significantly affect the performance of the model in this case.

##Bonus task: Implement a linear SVM classifier from scratch using Python and compare its performance with the scikit-learn implementation.

##Here is an example of how to implement a linear SVM classifier from scratch using the SMO algorithm:

````python
class SVM:
    def __init__(self, lr=0.001, num_iters=1000):
        self.lr = lr
        self.num_iters = num_iters
        self.w = None
        self.b = None
    
    def fit(self, X, y):
        n_samples, n_features = X.shape
        
        # initialize parameters
        self.w = np.zeros(n_features)
        self.b = 0
        
        # gradient descent
        for _ in range(self.num_iters):
            for idx, x_i in enumerate(X):
                condition = y[idx] * (np.dot(x_i, self.w) - self.b) >= 1
                if condition:
                    self.w -= self.lr * (2 * 0.01 * self.w)
                else:
                    self.w -= self.lr * (2 * 0.01 * self.w - np.dot(x_i, y[idx]))
                    self.b -= self.lr * y[idx]
    
    def predict(self, X):
        approx = np.dot(X, self.w) - self.b
        return np.sign(approx)
    
    def score(self, X, y):
        y_pred = self.predict(X)
        accuracy = np.sum(y_pred == y) / len(y)
        return accuracy
````

````python
my_svm = SVM()
my_svm.fit(X_train, y_train)
print("Custom SVM accuracy:", my_svm.score(X_test, y_test))

# compare with scikit-learn implementation
from sklearn.svm import SVC
sklearn_svm = SVC(kernel='linear', C=1)
sklearn_svm.fit(X_train, y_train)
print("Scikit-learn SVM accuracy:", sklearn_svm.score(X_test, y_test))
````
